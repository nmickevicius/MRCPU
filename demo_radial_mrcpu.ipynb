```python
# imports 

import numpy as np
import sigpy as sp
import matplotlib.pyplot as plt
import mrcpu_ncart
```

```python
# load data from .npz file

dataFile = 'data_radial_mrcpu.npz'
dd = np.load(dataFile)
traj = dd['traj']           # k-space trajectory (spokes,readout,2)
csm = dd['csm']             # coil sensitivity maps (coils,N,N)
ksp_E2_E3 = dd['ksp_E2_E3'] # data for aliased E2 and E3 pathways (coils,spokes,readout) 

ncoils = csm.shape[0]
N = csm.shape[1]
csm[np.abs(csm)==0] = 1.0E-16 
```

```python
# plot coil maps 

plt.figure()
for c in range(ncoils):
    plt.subplot(1,ncoils,c+1)
    plt.imshow(np.abs(csm[c,::,::]))
plt.show()
```

```python
# construct pathway phase modulation operator

# RF phase modulation schedule alternated in even/odd spokes 
phi = np.ones((2,traj.shape[0],traj.shape[1]),dtype=ksp_E2_E3.dtype)
phi[1,0::2,::] = np.exp(1j*np.pi)
```

```python
# retrospective acceleration 

# acceleration factor (R=1 corresponds to no acceleration)
#   Note: RF phase modulation schedule alternated in even/odd spokes
#         Use an odd integer acceleration factor to ensure proper 
#         phase modulation encoding is maintained
#   Note: not recommended to increase past 3
R = 3 

if (R%2 is not 1) or (R > 5):
    raise ValueError('R must be odd and less than or equal to 5')

# truncate the dataset
if R > 1:
    traj = traj[0::R,::,::]
    phi = phi[::,0::R,::]
    ksp_E2_E3 = ksp_E2_E3[::,0::R,::]
    
```

```python
# non-uniform Fourier transform operator 

F = sp.linop.NUFFT(ishape=(ncoils,2,N,N), coord=traj, oversamp=2.0)
```

```python
# ADMM MR-CPU reconstruction 

# set some parameters based on acceleration rate 
if R == 1:
    admmIter = 5 # number of ADMM iterations 
    cgIter = 10   # number of conjugate gradient iterations 
    lam = 2.0E-4  # \ell_1 wavelet regularization parameter
    rho = 1.0     # rho parameter for ADMM algorithm
elif R == 3: 
    admmIter = 6 # number of ADMM iterations 
    cgIter = 15   # number of conjugate gradient iterations 
    lam = 3.0E-4  # \ell_1 wavelet regularization parameter
    rho = 1.0     # rho parameter for ADMM algorithm
elif R == 5: 
    admmIter = 5 # number of ADMM iterations 
    cgIter = 15   # number of conjugate gradient iterations 
    lam = 4.0E-4  # \ell_1 wavelet regularization parameter
    rho = 1.0     # rho parameter for ADMM algorithm

# do the recon! (recon code is in mrcpu_ncart.py)
E = mrcpu_ncart.admm(ksp_E2_E3, F, phi, csm, admmIter, cgIter, rho, lam)
```

```python
# display the results 

plt.figure()
for e in range(E.shape[0]):
    img = np.abs(E[e,::,::])
    mn = np.mean(img,axis=None)
    sd = np.std(img,axis=None)
    vmin = 0
    vmax = mn + 3*sd
    plt.subplot(1,E.shape[0],e+1)
    plt.imshow(img,vmin=vmin,vmax=vmax,cmap='gray')
    plt.title('E%i Recon'%(e+2))
plt.show()
```
